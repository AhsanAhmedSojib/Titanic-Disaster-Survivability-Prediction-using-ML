```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
```

```python
test_data=pd.read_csv("test.csv")
train_data.shape
```

```python
print(pd.isnull(train_data).sum())
print(pd.isnull(test_data).sum())
```

```python
uSibSp=train_data['SibSp'].unique()
uSibSp
```

```python
uParch=train_data['Parch'].unique()
uParch
```

```python
train_data=train_data.drop(['PassengerId','Name','Ticket','Cabin'],axis=1)
train_data.head(3)
```

```python
test_data=test_data.drop(['Name','Ticket','Cabin'],axis=1)
test_data.tail(3)
```

```python
 train_data['Sex'].replace(['male','female'],[0,1],inplace=True)
 test_data['Sex'].replace(['male','female'],[0,1],inplace=True)
 train_data.head(3)
```

```python
train_data["Age"].replace(np.nan,30,inplace=True)
test_data["Age"].replace(np.nan,30,inplace=True)
print(pd.isnull(train_data).sum())
print(pd.isnull(test_data).sum())
```

```python
train_data["Embarked"]=train_data["Embarked"].fillna("S")
test_data["Embarked"]=test_data["Embarked"].fillna("S")

```

```python
print(pd.isnull(train_data).sum())
print(pd.isnull(test_data).sum())
```

```python
train_data["Fare"].fillna(train_data["Fare"].median(),inplace=True)
test_data["Fare"].fillna(test_data["Fare"].median(),inplace=True)
```

```python
print(pd.isnull(train_data).sum())
print(pd.isnull(test_data).sum())
```

```python
train_data.head()
```

```python
train_data['Fare']=train_data['Fare'].astype(int)
test_data['Fare']=test_data['Fare'].astype(int)
```

```python
train_data.head()
```

```python
test_data.tail()
```

```python
bins=[0,8,15,18,25,40,60,80]
names=['1','2','3','4','5','6','7']
train_data["Age"]=pd.cut(train_data['Age'],bins,labels=names)
test_data["Age"]=pd.cut(test_data['Age'],bins,labels=names)
train_data.head(3)
    
```

```python
embark_dummies_titanic=pd.get_dummies(train_data['Embarked'])
embark_dummies_titanic.drop(['S'],axis=1,inplace=True)
embark_dummies_test=pd.get_dummies(test_data['Embarked'])
embark_dummies_test.drop(['S'],axis=1,inplace=True)
train_data=train_data.join(embark_dummies_titanic)
test_data=test_data.join(embark_dummies_test)
train_data.drop(['Embarked'],axis=1,inplace=True)
test_data.drop(['Embarked'],axis=1,inplace=True)
train_data.head(3)
```

```python
print(test_data.shape)
```

```python
train_data.head()
```

```python
    test_data.head()
```

```python
from sklearn.model_selection import train_test_split
X=np.array(train_data.drop(['Survived'],1))
y=np.array(train_data['Survived'])
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
ids=test_data.PassengerId
```

```python
train_data['Family']= train_data["Parch"]+train_data["SibSp"]
train_data.head()
```

```python
test_data['Family']=test_data["Parch"]+test_data["SibSp"]
test_data.head()
```

```python
train_data['Large']=0
train_data.loc[train_data['Family']>=4,'Large']=1
train_data.loc[train_data['Family']>0,'Family']=1
train_data['Large']=train_data['Large'].astype(int)
train_data.head()
```

```python
test_data['Large']=0
test_data.loc[test_data['Family']>=4,'Large']=1
test_data.loc[test_data['Family']>0,'Family']=1
test_data['Large']=test_data['Large'].astype(int)
test_data.head()
```

```python
train_data=train_data.drop(['SibSp','Parch'],axis=1)
test_data=test_data.drop(['SibSp','Parch'],axis=1)
```

```python
uFamily=train_data['Family'].unique()
uFamily
```

```python

uLarge=train_data['Large'].unique()
uLarge
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
Y_pred=model.predict(test_data.drop('PassengerId',axis=1))
score_logreg=model.score(X_train,y_train)
print('Logistic Regression score TRAIN:',score_logreg)
score_logreg=model.score(X_test,y_test)
print('Logistic Regression score TEST:',score_logreg)
out_logreg=pd.DataFrame({'PassengerId':ids,'Survived':Y_pred})
print(out_logreg.head())

people=out_logreg.loc[out_logreg.Survived==1]["Survived"]
rate_people=0
if len(people)>0:
    rate_people=len(people)/len(out_logreg)
print("Logistic Regression % of people who Survived:",rate_people)
    
```

```python
#Support Vector Clustering(SVC)
from sklearn.svm import SVC, LinearSVC
model=SVC()
model.fit(X_train,y_train)
Y_pred=model.predict(test_data.drop('PassengerId',axis=1))
score_svc=model.score(X_train,y_train)
print('Support vector Clustering Score: ',score_svc)
score_svc=model.score(X_test,y_test)
print('Support vector Clustering Score: ',score_svc)

out_svc=pd.DataFrame({'PassengerId':ids,'Survived':Y_pred})
print(out_svc.head())

people = out_svc.loc[out_svc.Survived==1]["Survived"]
rate_people=0
if len(people)>0:
    rate_people=len(people)/len(out_svc)
print("Support Vector Clustering % of peope who survived:",rate_people)
```

```python
#K Nearest neighbour(KNN)
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(X_train,y_train)

Y_pred=model.predict(test_data.drop('PassengerId',axis=1))
score_knc=model.score(X_train,y_train)
print('K Nearest neighbor Classifier Score: ',score_knc)
score_knc=model.score(X_test,y_test)
print('K Nearest neighbor Classifier Score: ',score_knc)

out_knc=pd.DataFrame({'PassengerId':ids,'Survived':Y_pred})
print(out_knc.head())


people = out_knc.loc[out_knc.Survived==1]["Survived"]
rate_people=0
if len(people)>0:
    rate_people=len(people)/len(out_knc)
print("K Nearest neighbor Classifier % of peope who survived:",rate_people)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100)
model.fit(X_train,y_train)

Y_pred=model.predict(test_data.drop('PassengerId',axis=1))
score_rfc=model.score(X_train,y_train)
print('Random Forest Classifier Score: ',score_rfc)
score_rfc=model.score(X_test,y_test)
print('Random Forest Classifier Score: ',score_rfc)

out_rfc=pd.DataFrame({'PassengerId':ids,'Survived':Y_pred})
print(out_rfc.head())


people = out_rfc.loc[out_rfc.Survived==1]["Survived"]
rate_people=0
if len(people)>0:
    rate_people=len(people)/len(out_rfc)
print("Random Forest Classifier % of peope who survived:",rate_people)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
decision_tree=DecisionTreeClassifier(max_depth=5)
decision_tree.fit(X_train,y_train)

Y_pred=model.predict(test_data.drop('PassengerId',axis=1))
score_dtc=decision_tree.score(X_train,y_train)
print('Decision Tree Classifier Score Train: ',score_dtc)
score_dtc=decision_tree.score(X_test,y_test)
print('Decision Tree Classifier Score Test: ',score_dtc)

out_dtc=pd.DataFrame({'PassengerId':ids,'Survived':Y_pred})
print(out_dtc.head())


people = out_dtc.loc[out_dtc.Survived==1]["Survived"]
rate_people=0
if len(people)>0:
    rate_people=len(people)/len(out_dtc)
print("Decision Tree Classifier % of peope who survived:",rate_people)
```

```python
#compare Mechine learning Algorithms
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcdefaults()
fig,ax=plt.subplots()
people=('Logistic Regression',
       'Support Vector Clustering',
       'K Neighbors Classifier',
       'Random Forest Classifier',
       'Decision Tree Classifier')
y_pos=np.arange(len(people))
x=(score_logreg,score_svc,score_knc,score_rfc,score_dtc)
ax.barh(y_pos,x,align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()
ax.set_xlabel('Performance')
ax.set_title('Which one is the best algorithm?')
for i,v in enumerate(x):
    ax.text(v+1,i,str(v),color='black',va='center',fontweight='normal')
plt.show()
```
